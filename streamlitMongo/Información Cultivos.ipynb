```python
db = client["datasetCrops"]
collection = db["crops"]
parCultivo="Almonds, with shell"	
colec=collection.aggregate([{
        "$match":{"Item": parCultivo,
                  "Element": {"$in":["Area harvested","Production"]},
                  "Value":{'$ne': float('NaN')}}
    }, 
    {"$group":{"_id":["$Element","$Year"], 
               "totales": { "$sum": "$Value" }}}])
dfGrupo =pd.DataFrame(list(colec))
dfGrupo
```
